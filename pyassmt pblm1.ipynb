```python
def factorial(n):
    product = 1
    for i in range(n):
        #print(i,product)
        product *= (i + 1)
    return product
```

```python
print (factorial(3))
```

```python
def factorial_recursive(n):
    if n > 1:
        return n * factorial_recursive(n-1)
    else:
        return 1
```

```python
print(factorial_recursive(3))
```

```python
def compute_ratio(x, n):
    ratio = x**n / factorial(n)
    return ratio
```

```python
compute_ratio(2,3)
```

```python
def compute_sum(x,N):
    sum =1
    for i in range(N):
        sum += compute_ratio(x, i+1)
    return sum
```

```python
print(compute_sum(2, 3))
```
